**The loc() function is label based data selecting method which means that we have to pass the name of the row or column which we want to select. This method includes the last element of the range passed in it, unlike iloc(). loc() can accept the boolean data unlike iloc(). Many operations can be performed using the loc() method like**

```python
newdf.loc[0,0]= 124
#can also access with string colomn
```

```python
newdf.head()
```

```python
newdf.loc[2,3],[4,1]
```

```python
Salary.loc[(Salary['YearsExperience']<5)]
```

```python
#we can write more complex quaries using this "& keyword
Salary.loc[(Salary['YearsExperience']<5) & (Salary['Salary']>60000)]
```

**The iloc function in Python returns a view of the selected rows and columns from a Pandas DataFrame. This view can be used to access, modify, or delete the selected data. The returned view is a Pandas DataFrame or Series, depending on the number of rows or columns selected.**

```python
Salary.iloc[[2,4]]
```

```python
#deleting the YearsExperience colomn
del Salary['YearsExperience']
```
